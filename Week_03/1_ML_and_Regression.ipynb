```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, MinMaxScaler
```

```python
# Read the regression training dataset
df = pd.read_csv('regression_training_dataset.csv')
df.shape
df.head()
```

```python
x = df.drop(columns=['PerformanceScore'])
y = df['PerformanceScore']
```

```python
y_stratified_bins = pd.qcut(y, q=5, labels=False)
```

```python
X_train_val, X_test, y_train_val, y_test = train_test_split(
    x, y,
    test_size=0.20,
    random_state=42,
    stratify=y_stratified_bins
)
```

```python
y_train_val_bins = pd.qcut(y_train_val, q=5, labels=False)

X_train, X_val, y_train, y_val = train_test_split(
    X_train_val, y_train_val,
    test_size=0.25,
    random_state=42,
    stratify=y_train_val_bins
)
```
