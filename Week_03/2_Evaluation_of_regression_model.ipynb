MSE:<br>
-> evaluate the accuracy of linear regression<br>
-> The goal is to minimize the MSE<br>
-> As a lower MSE indicates that the model's predictions are closer to the actual values.<br>
    
    math formula:
    MSE = (1/n) * Σ(y_i - ŷ_i)²

    Table 
    Hours Studied | Actual Score | Predicted Score | (y - ŷ)²
        1         | 2            | 2.5             | 0.25
        2         | 4            | 4               | 0
        3         | 5            | 5.5             | 0.25
        4         | 4.5          | 7               | 6.25
        5         | 7            | 8.5             | 2.25

        MSE = (0.25 + 0 + 0.25 + 6.25 + 2.25) / 5 = 9 / 5 = 1.8


What if MSE is 0?<br>
If the Mean Squared Error (MSE) is 0, it indicates that the model's predictions are perfect and exactly match the actual values for all data points. In other words, there is no error in the predictions, and the model has achieved the best possible accuracy. However, in practice, an MSE of 0 is extremely rare and may suggest overfitting if it occurs on training data, meaning the model may not generalize well to new, unseen data.
