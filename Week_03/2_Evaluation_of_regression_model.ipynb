MSE:<br>
-> evaluate the accuracy of linear regression<br>
-> The goal is to minimize the MSE<br>
-> As a lower MSE indicates that the model's predictions are closer to the actual values.<br>
    
    math formula:
    MSE = (1/n) * Σ(y_i - ŷ_i)²

y means the actual value, ŷ means the predicted value, and n is the number of data points.

    Table 
    Hours Studied | Actual Score | Predicted Score | (y - ŷ)²
        1         | 2            | 2.5             | 0.25
        2         | 4            | 4               | 0
        3         | 5            | 5.5             | 0.25
        4         | 4.5          | 7               | 6.25
        5         | 7            | 8.5             | 2.25

        MSE = (0.25 + 0 + 0.25 + 6.25 + 2.25) / 5 = 9 / 5 = 1.8


What if MSE is 0?<br>
If the Mean Squared Error (MSE) is 0, it indicates that the model's predictions are perfect and exactly match the actual values for all data points. In other words, there is no error in the predictions, and the model has achieved the best possible accuracy. However, in practice, an MSE of 0 is extremely rare and may suggest overfitting if it occurs on training data, meaning the model may not generalize well to new, unseen data.


Penalises larger errors more than smaller ones<br>
MSE is sensitive to outliers because it squares the differences between predicted and actual values.<br>
If diff is 40 then it squared is 1600.

**Root Mean Squared Error (RMSE)**<br>
Formula: RMSE = √MSE<br>
RMSE is the square root of the Mean Squared Error (MSE) and provides a measure of the average magnitude of the errors in the same units as the target variable. It is often preferred over MSE because it is more interpretable and gives a sense of how far off predictions are from actual values on average.

Mean Absolute Error (MAE)<br>

    Math Formula: MAE = (1/n) * Σ|y_i - ŷ_i|
MAE is the average of the absolute differences between predicted and actual values. It is less sensitive to outliers compared to MSE, as it does not square the errors.

R-squared (R²) Coefficient of Determination<br>

    SS_res = Σ(y_i - ŷ_i)² (Residual Sum of Squares)<br>
    SS_tot = Σ(y_i - ȳ)² (Total Sum of Squares)<br>
    Math Formula: R² = 1 - (SS_res / SS_tot)

R² is a statistical measure that represents the proportion of the variance in the dependent variable that is predictable from the independent variable(s). It ranges from 0 to 1, where 1 indicates a perfect fit and 0 indicates no linear relationship.

**Core Intuition**<br>
 R² acts as a standardized measure of how well the model's predictions match the actual data. A higher R² value indicates that a larger proportion of the variance in the dependent variable is explained by the model, suggesting better predictive performance. Conversely, a lower R² value suggests that the model does not capture much of the variability in the data, indicating poorer predictive performance.

```python
import numpy as np
import pandas as pd
```

```python
df = pd.DataFrame({
    'Actual': [2, 4, 5, 4.5, 7],
    'Predicted': [2.5, 4, 5.5, 7, 8.5]
})
```

```python
#Calculation R-squared (R²) Coefficient of Determination
SS_res = np.sum((df['Actual'] - df['Predicted']) ** 2)
SS_tot = np.sum((df['Actual'] - np.mean(df['Actual'])) ** 2)

R_squared = 1 - (SS_res / SS_tot)

R_squared
```

**K-Fold Cross-Validation**<br>

**The Core Concept**<br>

if we evaluate our model using only a single, fixed Train/Test split, our metrics are susceptible to<br>
sampling bias. By random chance the holdout test set might contain uncharaceteristically easy or<br>
unusually difficult data points, leading to misleadngly high or low performance scores.

k-Fold Cross-Validation: provides a more stable, reliable approach to evaluation by following a structred rotation.

**Why?**<br>
 -> Reduce Overfitting<br>
 -> Efficient use of data<br>
 -> Reliable Performance Estimation<br>

**How?**<br>
 -> The model is trained on k-1 folds and tested on the remaining fold.<br>
 -> The process is repeated K-Times with each fold serving as the test set once.<br>


        k1         k2         k3         k4
[16, 22, 31, 47, 55, 63, 71, 88, 95, 100, 111, 124] <br>

Iteration 1: Train on k2, k3, k4, Test on k1<br>
Iteration 2: Train on k1, k3, k4, Test on k2<br>
Iteration 3: Train on k1, k2, k4, Test on k3<br>
Iteration 4: Train on k1, k2, k3, Test on k4<br>


**Main Goal**<br>
Average performance of all iterations of training and tesing.

**<h4> The concept of Bias and Variance in Machine Learning</h4>**<br>
The total expected prediction error for a model can be decomposed into three components: bias, variance, and irreducible error. This decomposition is known as the bias-variance tradeoff.

Expected Error = Bias² + Variance + Irreducible Error

Where irreducible error represents the noise inherent in the data that cannot be reduced by any model.

**High Bias**<br>
When a model is too simple it underfits the data, this means the model doesn't capture important patterns in the data, leading to poor performance on both the training and test sets. 

**Low Bias**<br>
Highly accurate or low error(almost none) in training data.

**Variance**<br>
It is the gap between the model's performance on the training data and its performance on unseen test data. High variance indicates that the model is sensitive to small fluctuations in the training data, leading to overfitting. Low variance suggests that the model generalizes well to new data.<br>
Example:
Training Error: 20<br>
Test Error: 70<br>
Variance = Test Error - Training Error = 70 - 20 = 50 (High Variance)<br>

**For ideal model**<br>
There should be a balance between bias and variance. The goal is to find a model that has low bias and low variance, which means it captures the underlying patterns in the data without being overly sensitive to noise. This balance is crucial for achieving good generalization performance on unseen data.
