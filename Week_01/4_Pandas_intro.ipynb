```python
df_students = pd.read_csv("students_pandas_day4.csv")
df_students.head()
```

```python
# check the structural layout and the data types of the dataframe
df_students.info()
```

```python
# Generate summary statistics for numerical scores

print("---- Summary Statistics for Numerical Scores ---")
print(df_students.describe())
```

```python
print('\nShape of the DataFrame (Rows, Columns):', df_students.shape)
```

```python
names_column = df_students['Name']
grades_df = df_students[['Name', 'Maths', 'Science']]
```

```python
computing_filter = df_students['Course'] == 'Computing' 
computing_students = df_students[computing_filter]
```

```python
# Advanced Filtering: Students with Maths score >= 80 and attendance >= 90
high_achievers = df_students[(df_students['Maths'] >= 80) & (df_students['Attendance'] >= 90)]
```

```python
print("--- Computing Students ---")
print(computing_students[['Name', 'Course']])

print("\n--- High Achievers ---")
print(high_achievers[['Name', 'Maths', 'Attendance']])
```

```python
# sorting
top_attendance = df_students.sort_values(by='Attendance', ascending=False)
print("\n--- Students Sorted by Attendance ---")
print(top_attendance[['Name', 'Attendance']].head())
```

```python
# accessing cells using label-based indexing with .loc
student_course_loc = df_students.loc[1, 'Course']
print("\nCourse of the second student (using .loc):", student_course_loc)
```

```python
# Accessing cells using position-based indexing with .iloc
student_maths_iloc = df_students.iloc[1:3, 5]  # Row 1, Column 5 (Maths score)
print("Maths score of the second student (using .iloc):\n",student_maths_iloc)
```

```python
# final three records of the dataset
print("\n--- Last Three Records of the Dataset ---")
print(df_students.tail(3))
```

```python
# Extract all the female students who are female and belong to Business course
female_business_students = df_students[(df_students['Gender'] == 'F') & (df_students['Course'] == 'Business')]
print("\n---Female Business Students---")
print(female_business_students[['Name','Gender','Course']])
```
