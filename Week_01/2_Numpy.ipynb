```python
import numpy as np
```

**creating a 1D array from a list**

```python
sales_list = [120, 150, 300, 250, 90]
sales_array = np.array(sales_list)
print(sales_list)
print(sales_array)
```

**creating a 2D array from a list of lists**

```python
# features: [age, income_in_k, credit_score]

customer_matrix = np.array([[25, 50, 710],
                            [41, 95, 780],
                            [32, 65, 620]])

print(customer_matrix)
print("Customer matrix shape:", customer_matrix.shape)
```

**creating a matrix of zeros**

```python
zero_matrix = np.zeros((3, 4))
print(zero_matrix)
```

**Arange function to create a sequenced array**

```python
sequenced_array = np.arange(0,10,2)
print(sequenced_array)
```

**Slicing a 1D array**

```python
traffic = np.array([500, 600, 550, 700, 650, 900, 1100])
weekend_traffic = traffic[-2:]
print(weekend_traffic)
```

**Setup a 2D array to represent sensor readings**

```python
sensors = np.array([
            [12.1,14.3,15.2],#sensor 0
            [11.5,13.8,14.9],#sensor 1
            [10.8,12.5,13.7],#sensor 2
            [9.5,11.2,12.4] #sensor 3
        ])
```

```python
sensors
```

```python
specific_reading = sensors[2, 1]
print(specific_reading)
```

```python
top_sensors = sensors[0:2, :]
print(top_sensors)
```
