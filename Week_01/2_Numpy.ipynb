```python
import numpy as np
```

**creating a 1D array from a list**

```python
sales_list = [120, 150, 300, 250, 90]
sales_array = np.array(sales_list)
print(sales_list)
print(sales_array)
```

**creating a 2D array from a list of lists**

```python
# features: [age, income_in_k, credit_score]

customer_matrix = np.array([[25, 50, 710],
                            [41, 95, 780],
                            [32, 65, 620]])

print(customer_matrix)
```

**creating a matrix of zeros**

```python
zero_matrix = np.zeros((3, 4))
print(zero_matrix)
```

**Arange function to create a sequenced array**

```python
sequenced_array = np.arange(0,10,2)
print(sequenced_array)
```
