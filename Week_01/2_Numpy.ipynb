**creating a 1D array from a list**

**Slicing a 1D array**

```python
traffic = np.array([500, 600, 550, 700, 650, 900, 1100])
weekend_traffic = traffic[-2:]
print(weekend_traffic)
```

**Setup a 2D array to represent sensor readings**

```python
sensors = np.array([
            [12.1,14.3,15.2],#sensor 0
            [11.5,13.8,14.9],#sensor 1
            [10.8,12.5,13.7],#sensor 2
            [9.5,11.2,12.4] #sensor 3
        ])
```

```python
sensors
```

```python
specific_reading = sensors[2, 1]
print(specific_reading)
```

```python
top_sensors = sensors[0:2, :]
print(top_sensors)
```

```python
final_reading = sensors[:, 2]
print(final_reading)
```

**Reshaping Example**

```python
flat_temps = np.array([22, 24, 19, 21, 23, 20, 25, 18, 22, 24, 18, 21]) # temperatures in celsius.
# reshape into 4 day, 3 readings per day
reshaped_temps = flat_temps.reshape(4, 3)
print(reshaped_temps)
```

**Broadcasting Example**

```python
farenheit_temps = (reshaped_temps * 9/5) + 32
print(farenheit_temps)
```

```python
scores = np.array([[85, 90, 78],
                   [92, 88, 95],
                   [80, 85, 82]])
```

```python
# global statistics
print("Overall Highest Score:", np.max(scores))
print("Overall Lowest Score:", np.min(scores))
print("Overall Average Score:", np.mean(scores).round(2))
```

```python
# Statistics by axis
# Average scores for each subject (axis 0)
subject_averages = np.mean(scores, axis=0).round(2)
print("Average Score for Each Subject:", subject_averages)

# Average scores for each student (axis 1)
student_averages = np.mean(scores, axis=1).round(2)
print("Average Score for Each Student:", student_averages)
```

**Python list vs Numpy array performance comparison**

```python
import time
size = 1_000_000
python_list = list(range(size))
numpy_array = np.arange(size)
```

```python
start_time = time.time()
python_list_result = [x * 2 for x in python_list]
python_duration = time.time() - start_time
print(f"Python list processing time: {python_duration:.4f} seconds")

start_time = time.time()
numpy_result = numpy_array * 2
numpy_duration = time.time() - start_time
print(f"Numpy array processing time: {numpy_duration:.4f} seconds")

print(f"Numpy is approximately {python_duration / numpy_duration:.2f} times faster than Python list for this operation.")
```
